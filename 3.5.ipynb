```python
from PIL import Image
import matplotlib.pyplot as plt

def resize_image(image, scale_factor):
    """Resizes the image to a specified scale factor using Lanczos resampling."""
    size = (int(image.width * scale_factor), int(image.height * scale_factor))
    return image.resize(size, Image.Resampling.LANCZOS)

def display_images(images, scale_factors):
    """Displays a list of images in a single row with titles indicating the scale factor."""
    fig, axs = plt.subplots(1, len(images), figsize=(20, 5))
    for ax, img, scale in zip(axs, images, scale_factors):
        ax.imshow(img)
        ax.set_title(f'Scale Factor: {scale}')
        ax.axis('off')
    plt.tight_layout()
    plt.show()

def main(image_path):
    original_image = Image.open(image_path)

    scale_factors = [1, 0.5, 0.25, 0.125]

    scaled_images = [resize_image(original_image, scale) for scale in scale_factors]

    display_images(scaled_images, scale_factors)


image_path = '20200713184622902.jpeg'


if __name__ == "__main__":
    main(image_path)
```
