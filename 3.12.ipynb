```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def load_image(image_path):
    return cv2.imread(image_path)

def filp_image(image,axis=0):
    return cv2.flip(image,axis)

def adjust_contrast(image,alpha):
    return cv2.convertScaleAbs(image,alpha=alpha)

def power_transformation(image,gamma):
    lookUpTable = np.array([((i/255)**gamma)*255 for i in np.arange(0,256)]).astype("uint8")
    return cv2.LUT(image,lookUpTable)

def segmentation_grayscale(image):
    gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    _,segmented=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
    return segmented

image_path = '2.jpg'
image = load_image(image_path)

flipped = filp_image(image,1)
contrast_adjusted = adjust_contrast(flipped,1.5)
power_transformed = power_transformation(contrast_adjusted,0.5)
segmentation_gray = segmentation_grayscale(power_transformed)

plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.title('Flipped Image')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(cv2.cvtColor(contrast_adjusted, cv2.COLOR_BGR2RGB))
plt.title('Contrast Adjusted Image')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(cv2.cvtColor(power_transformed, cv2.COLOR_BGR2RGB))
plt.title('Power Transformed Image')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(segmentation_gray, cmap='gray')
plt.title('Segmented Grayscale Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np

def load_image(image_path):

    return cv2.imread(image_path)

def convert_to_grayscale(image):

    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def power_transformation(image, gamma):

    lookUpTable = np.array([((i / 255.0) ** gamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
    return cv2.LUT(image, lookUpTable)


image_path = '2.jpg'  
image = load_image(image_path)
gray_image = convert_to_grayscale(image)

gamma_value = 0.5  
power_transformed_image = power_transformation(gray_image, gamma_value)


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(power_transformed_image, cmap='gray')  
plt.title(f'Power Transformed Image (gamma={gamma_value})')
plt.axis('off')

plt.show()
```
