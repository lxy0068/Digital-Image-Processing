```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_image_info(image_path):
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  

    plt.imshow(image_rgb)
    plt.title('Original Image')
    plt.axis('off')
    plt.show()

    height, width, channels = image.shape
    print(f"Dimensions: Height = {height}, Width = {width}, Channels = {channels}")
    print(f"Data Type: {image.dtype}")

    x, y = 10, 10
    pixel_values = image[y, x]
    print(f"Pixel values at ({x}, {y}): {pixel_values}")

image_path = '1.jpg'  
display_image_info(image_path)
```

```python
def convert_to_grayscale(image_path):
    color_image = cv2.imread(image_path)

    grayscale_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

    plt.imshow(grayscale_image, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')
    plt.show()

    return grayscale_image

image_path = '1.jpg'  
convert_to_grayscale(image_path)
```

```python
def invert_grayscale_image(image_path):
    color_image = cv2.imread(image_path)

    grayscale_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

    inverted_image = 255 - grayscale_image

    plt.imshow(inverted_image, cmap='gray')
    plt.title('Inverted Grayscale Image')
    plt.axis('off')
    plt.show()

    return inverted_image


image_path = '1.jpg'  
invert_grayscale_image(image_path)
```

```python
def digitize_and_decompose_rgb(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  

    R, G, B = cv2.split(image)

    plt.figure(figsize=(12, 4))
    for i, channel in enumerate([R, G, B], start=1):
        plt.subplot(1, 3, i)
        plt.imshow(channel, cmap='gray')
        plt.title(f'Channel {i} ({"RGB"[i-1]})')
        plt.axis('off')
    
    plt.show()

    sample_pixels = [(10, 10), (20, 20), (30, 30)]  
    for x, y in sample_pixels:
        print(f"Pixel at ({x}, {y}): R={R[y, x]}, G={G[y, x]}, B={B[y, x]}")

image_path = '1.jpg'  
digitize_and_decompose_rgb(image_path)
```

```python
# 归一化
import cv2
import numpy as np
import matplotlib.pyplot as plt

def normalize_and_convert_to_grayscale(image_path):
    image = cv2.imread(image_path)

    normalized_image = cv2.normalize(image, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)

    grayscale_image = cv2.cvtColor(normalized_image, cv2.COLOR_RGB2GRAY)

    plt.imshow(grayscale_image, cmap='gray')
    plt.title('Normalized Grayscale Image')
    plt.axis('off')
    plt.show()

    return grayscale_image

image_path = '1.jpg'  
normalize_and_convert_to_grayscale(image_path)
```
