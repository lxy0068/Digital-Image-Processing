```python
from PIL import Image
import matplotlib.pyplot as plt

# 加载原始图像
original_image_path = "image_without_edge.jpg"
image = Image.open(original_image_path)

# 设置顶部和右侧边缘的大小
top_edge = 50
right_edge = 50

# 计算新图像的宽度和高度
new_width = image.width + right_edge
new_height = image.height + top_edge

# 创建一个新的白色图像
new_image = Image.new("RGB", (new_width, new_height), "white")

# 将原始图像粘贴到新图像中，考虑顶部边缘偏移
new_image.paste(image, (0, top_edge))

# 保存新图像
new_image_path = "image_adding_edge.jpg"
new_image.save(new_image_path)

# 显示新图像
plt.figure(figsize=(8, 6))
plt.imshow(new_image)
plt.axis('on')  # 显示坐标轴
plt.title('Image with Added White Edges')  # 设置标题
plt.show()
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# 设置已处理并添加白色边缘的图像的路径
new_image_with_corners_path = "image_without_corners.png"

# 从文件中加载图像
image_with_corners = Image.open(new_image_with_corners_path)

# 定义需要添加的新白色边缘的尺寸
top_edge = 50
right_edge = 50

# 计算图像新的尺寸
new_width = image_with_corners.width + right_edge
new_height = image_with_corners.height + top_edge

# 创建一个具有新尺寸的白色背景新图像
new_image_with_corners = Image.new("RGB", (new_width, new_height), "white")

# 将原始处理过的图像粘贴到新图像上，通过偏移来创建顶部边缘
# 这也通过在粘贴图像的右侧留出空间来创建右边缘
new_image_with_corners.paste(image_with_corners, (0, top_edge))

# 为了保留原图像，将添加了边缘的新图像保存为一个新的文件名
modified_image_path = "image_with_added_edges.png"
new_image_with_corners.save(modified_image_path)

# 为展示加载图像
image_to_display = Image.open(modified_image_path)

plt.figure(figsize=(8, 6))
plt.imshow(image_to_display)
plt.axis('on') 
plt.title('Image with Added White Edges') 
plt.show()
```
